### Plotting the feature correlation matrix of the Wine dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

url = "https://raw.githubusercontent.com/basilhan/datasets/master/wine.data"
names =  ['Class', 'Alcohol', 'MalicAcid', 'Ash', 'AlcalinityOfAsh', 'Magnesium', 'TotalPhenols', 
         'Flavanoids', 'NonflavanoidPhenols', 'Proanthocyanins', 'ColorIntensity', 
         'Hue', 'OD280/OD315', 'Proline']
data = pd.read_csv(url, names=names)
data.head()
```

```python
data.shape
```

```python
X = data.drop('Class', axis=1)
X.head()
```

```python
import numpy as np
cm = np.corrcoef(X.values.T)
sns.set(font_scale=1.3)
hm = sns.heatmap(
    cm,
    cbar=True,
    annot=True,
    square=False,
    fmt='.2f',
    annot_kws={'size': 10},
    yticklabels=list(X.columns.values),
    xticklabels=list(X.columns.values)
)
plt.show()
```
