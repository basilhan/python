### Scatter plotting the Wine dataset

```python
import pandas as pd
%matplotlib inline

url = "https://raw.githubusercontent.com/basilhan/datasets/master/wine.data"
names =  ['Class', 'Alcohol', 'MalicAcid', 'Ash', 'AlcalinityOfAsh', 'Magnesium', 'TotalPhenols', 
         'Flavanoids', 'NonflavanoidPhenols', 'Proanthocyanins', 'ColorIntensity', 
         'Hue', 'OD280/OD315', 'Proline']
data = pd.read_csv(url, names=names)
data.head()
```

```python
data.shape
```

```python
y = data['Class']
y.head()
```

```python
X = data.drop('Class', axis=1)
X.head()
```

#### Plotting 2 features

```python
import matplotlib.pyplot as plt

plt.scatter(X[y==1]['Flavanoids'], X[y==1]['NonflavanoidPhenols'], label='Class 1', c='red')
plt.scatter(X[y==2]['Flavanoids'], X[y==2]['NonflavanoidPhenols'], label='Class 2', c='blue')
plt.scatter(X[y==3]['Flavanoids'], X[y==3]['NonflavanoidPhenols'], label='Class 3', c='lightgreen')

plt.legend()
plt.xlabel('Flavanoids')
plt.ylabel('NonflavanoidPhenols')

plt.show()
```

#### Plotting all features in a scatter plot matrix

```python
import seaborn as sns
sns.set()

sns.pairplot(data, hue='Class')
```
