### Box plotting the Wine dataset

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
%matplotlib inline

url = "https://raw.githubusercontent.com/basilhan/datasets/master/wine.data"
names =  ['Class', 'Alcohol', 'MalicAcid', 'Ash', 'AlcalinityOfAsh', 'Magnesium', 'TotalPhenols', 
         'Flavanoids', 'NonflavanoidPhenols', 'Proanthocyanins', 'ColorIntensity', 
         'Hue', 'OD280/OD315', 'Proline']
data = pd.read_csv(url, names=names)
data.head()
```

```python
data.shape
```

```python
X = data.drop('Class', axis=1)
X.head()
```

```python
plt.boxplot(X.values);
```

It appears that the features contain values of incomparable ranges. A descriptive statistical summary confirms this.

```python
X.describe()
```

It would therefore be a good idea to standardize the values to zero mean and unit variance in order to compare the distribution of the features,.

```python
scaler = StandardScaler()
scaler.fit(X)
```

```python
plt.figure(figsize=(14,8))
plt.boxplot(scaler.transform(X))
plt.xticks(np.arange(14), names, rotation=70)
plt.tick_params(labelsize=12);
```
